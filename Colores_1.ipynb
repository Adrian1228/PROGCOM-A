```python
# importar los módulos
import tkinter
import random

# lista de posibles colores
colores = ['Red','Blue','Green','Pink','Black',
           'Yellow','Orange','White','Purple','Brown']
puntaje = 0

# tiempo restante del juego, inicialmente 30 segundos
tiempo_restante = 30

# función que iniciará el juego
def iniciar_juego(evento):
    global puntaje
    if tiempo_restante == 30:
        # reiniciar puntaje al comenzar
        puntaje = 0
        etiqueta_puntaje.config(text="Puntaje: 0")
        # iniciar el contador regresivo
        cuenta_regresiva()

    # ejecutar la función para elegir el siguiente color
    siguiente_color()

# función para elegir y mostrar el siguiente color
def siguiente_color():
    global puntaje
    global tiempo_restante

    # si el juego sigue activo
    if tiempo_restante > 0:
        # activar la caja de texto
        entrada.focus_set()

        # verificar la respuesta ANTES de cambiar los colores
        if entrada.get().lower() == etiqueta_color.cget("fg").lower():
            puntaje += 1

        # limpiar la caja de texto
        entrada.delete(0, tkinter.END)

        # mezclar los colores
        random.shuffle(colores)

        # cambiar el color y el texto
        etiqueta_color.config(fg=str(colores[1]), text=str(colores[0]))

        # actualizar el puntaje
        etiqueta_puntaje.config(text="Puntaje: " + str(puntaje))

# función del temporizador regresivo
def cuenta_regresiva():
    global tiempo_restante

    # si el juego sigue activo
    if tiempo_restante > 0:
        # disminuir el tiempo
        tiempo_restante -= 1
        # actualizar la etiqueta del tiempo
        etiqueta_tiempo.config(text="Tiempo restante: " + str(tiempo_restante))
        # ejecutar la función de nuevo después de 1 segundo
        etiqueta_tiempo.after(1000, cuenta_regresiva)
    else:
        etiqueta_color.config(text="¡Tiempo terminado!", fg="black")

# Código principal (ventana del juego)

ventana = tkinter.Tk()
ventana.title("JUEGO DE COLORES")
ventana.geometry("375x200")

instrucciones = tkinter.Label(ventana, text="¡Escribe el color del texto, no la palabra!",
                              font=('Helvetica', 12))
instrucciones.pack()

etiqueta_puntaje = tkinter.Label(ventana, text="Presiona Enter para empezar",
                                 font=('Helvetica', 12))
etiqueta_puntaje.pack()

etiqueta_tiempo = tkinter.Label(ventana, text="Tiempo restante: " + str(tiempo_restante),
                                font=('Helvetica', 12))
etiqueta_tiempo.pack()

etiqueta_color = tkinter.Label(ventana, font=('Helvetica', 60))
etiqueta_color.pack()

entrada = tkinter.Entry(ventana)
ventana.bind('<Return>', iniciar_juego)
entrada.pack()
entrada.focus_set()

ventana.mainloop()
```
