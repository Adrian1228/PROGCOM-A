```python
##Problema: Se necesita calcular el promedio, mediana, varianza y desviación estándar de las notas de un grupo de estudiantes, hacerlo a mano tomo tiempo y pueden haber errores. Por este mmotivo, he creado la solución para esto.

import statistics as stats
def notas():
    print("Análisis de Notas")
    print()
    print("Ingresa las notas separadas por coma  (1.2,3.4,2.8,4.6)")

    entrada = input("Notas: ")

    nota = [float(x) for x in entrada.split(",")]

    print(f"\nNotas ingresadas: {nota}\n")

    promedio = stats.mean(nota)
    mediana = stats.median(nota)
    varianza = stats.variance(nota)
    desviacion = stats.stdev(nota)

    print(f"Promedio: {promedio:.2f}")
    print(f"Mediana: {mediana:.2f}")
    print(f"Varianza: {varianza:.2f}")
    print(f"Desviación estándar: {desviacion:.2f}")

notas()
```

```python
import tkinter as tk
from tkinter import messagebox
import statistics as stats

def calcular():
    try:
        notas = [float(x) for x in entrada.get().split(",")]

        promedio = stats.mean(notas)
        mediana = stats.median(notas)
        varianza = stats.variance(notas)
        desviacion = stats.stdev(notas)
        moda = stats.mode(notas)
        rango = max(notas) - min(notas)
        mayor = max(notas)
        menor = min(notas)

        # Reprueba si la nota es menor a 3.0
        aprobados = len([n for n in notas if n >= 3.0])
        reprobados = len([n for n in notas if n < 3.0])

        resultado.set(
            f" Resultados del análisis:\n\n"
            f"Promedio: {promedio:.2f}\n"
            f"Mediana: {mediana:.2f}\n"
            f"Moda: {moda:.2f}\n"
            f"Varianza: {varianza:.2f}\n"
            f"Desviación estándar: {desviacion:.2f}\n"
            f"Rango: {rango:.2f}\n"
            f"Nota más alta: {mayor:.2f}\n"
            f"Nota más baja: {menor:.2f}\n"
            f"Aprobados (≥3.0): {aprobados}\n"
            f"Reprobados (<3.0): {reprobados}"
        )
    except:
        messagebox.showerror("Error", " Ingresa solo números separados por coma.\nEjemplo: 3.5,4.0,2.8,5.0")

def limpiar():
    entrada.delete(0, tk.END)
    resultado.set("")

# --- Interfaz ---
ventana = tk.Tk()
ventana.title("Analizador de Notas")
ventana.geometry("420x480")
ventana.config(bg="#E8EAF6")

tk.Label(
    ventana, text="ANÁLISIS DE NOTAS", bg="#3F51B5", fg="white",
    font=("Arial", 14, "bold"), pady=10
).pack(fill=tk.X)

tk.Label(
    ventana, text="Ingresa las notas separadas por coma:",
    bg="#E8EAF6", font=("Arial", 11)
).pack(pady=(15,5))

entrada = tk.Entry(ventana, width=40, font=("Arial", 11))
entrada.pack(pady=5)

tk.Button(
    ventana, text="Calcular", bg="#4CAF50", fg="white",
    font=("Arial", 11, "bold"), width=12, command=calcular
).pack(pady=8)

tk.Button(
    ventana, text="Limpiar", bg="#F44336", fg="white",
    font=("Arial", 11, "bold"), width=12, command=limpiar
).pack(pady=5)

resultado = tk.StringVar()
tk.Label(
    ventana, textvariable=resultado, bg="#E8EAF6",
    justify="left", font=("Consolas", 11), wraplength=380
).pack(pady=20)

tk.Label(
    ventana, text="Desarrollado por: [Tu nombre aquí]",
    bg="#E8EAF6", fg="#555", font=("Arial", 9, "italic")
).pack(side="bottom", pady=10)

ventana.mainloop()

```
