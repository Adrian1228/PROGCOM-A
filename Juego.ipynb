```
import tkinter as tk  # Interfaz Gráfica
import pygame  # Sonido
from PIL import Image, ImageTk  # Para imágenes

pygame.mixer.init()

def audio(archivo):
    pygame.mixer.music.load(archivo)
    pygame.mixer.music.play()

def iniciar_juego():
    audio("D:/INTRO.wav")
    resultado.config(text="1", font=("Arial", 40), fg="white", bg="#6a0dad")
    juego.after(3000, lambda: mostrar_pregunta(0))  # Espera 3 segundos

def mostrar_pregunta(indice):
    resultado.config(text="")
    pregunta, imagen, opciones, correcta, audio_correcto = preguntas[indice]

    pregunta_label.config(text=pregunta)
    img = Image.open(imagen)
    img = img.resize((150, 150))
    foto = ImageTk.PhotoImage(img)
    imagen_label.config(image=foto)
    imagen_label.image = foto

    for widget in opciones_frame.winfo_children():
        widget.destroy()

    for o in opciones:
        tk.Button(opciones_frame, text=o, width=10, height=2,
                  command=lambda opcion=o, i=indice: verificar_respuesta(opcion, correcta, i, audio_correcto)).pack(side="left", padx=5)

def verificar_respuesta(opcion, correcta, indice, audio_correcto):
    for widget in opciones_frame.winfo_children():
        widget.destroy()

    if opcion == correcta:
        audio(audio_correcto)
        if indice + 1 < len(preguntas):
            resultado.config(text=str(indice + 2), font=("Arial", 40), fg="white", bg="#6a0dad")
            juego.after(3000, lambda: mostrar_pregunta(indice + 1))
        else:
            audio("D:/ganaste.wav")
            mostrar_mensaje_final("CONGRATULATIONS! \nYa sabes algunos animales en inglés,\nsigue practicando")
    else:
        audio("D:/perdiste.wav")
        mostrar_mensaje_final("Oh no! ❌\nEsa no era, pero tranquilo,\nde los errores se aprende")

def mostrar_mensaje_final(mensaje):
    resultado.config(text=mensaje, font=("Arial", 20), fg="yellow", bg="#6a0dad")
    pregunta_label.config(text="")
    imagen_label.config(image="")

    for widget in opciones_frame.winfo_children():
        widget.destroy()

    reiniciar_btn = tk.Button(opciones_frame, text="REINICIAR", font=("Arial", 16), bg="green", fg="white", command=reiniciar_juego)
    reiniciar_btn.pack(pady=10)

def reiniciar_juego():
    resultado.config(text="Bienvenidos a AnimalQuiz", font=("Arial", 24), fg="white", bg="#6a0dad")
    pregunta_label.config(text="")
    imagen_label.config(image="")
    for widget in opciones_frame.winfo_children():
        widget.destroy()

preguntas = [
    ("¿Cómo se escribe vaca en inglés?", "D:/VACA.png", ["Cow", "Caw", "Cew"], "Cow", "D:/Vaca.wav"),
    ("¿Cómo se escribe pato en inglés?", "D:/PATO1.JPEG", ["dock", "duck", "dack"], "duck", "D:/Pato.wav"),
    ("¿Cómo se escribe gato en inglés?", "D:/GATO.JPG", ["Cat", "Cot", "Cut"], "Cat", "D:/Gato.wav"),
    ("¿Cómo se escribe oveja en inglés?", "D:/OVEJA.png", ["shipishapa", "shep", "sheep"], "sheep", "D:/Oveja.wav")
]

juego = tk.Tk()
juego.title("Bienvenidos a AnimalQuiz")
juego.geometry("500x500")
juego.configure(bg="#6a0dad")

resultado = tk.Label(juego, text="Bienvenidos a AnimalQuiz", font=("Arial", 24), fg="white", bg="#6a0dad")
resultado.pack(pady=20)

tk.Button(juego, text="PLAY", font=("Arial", 20), bg="green", fg="white",
          command=iniciar_juego).pack(pady=10)

pregunta_label = tk.Label(juego, text="", font=("Arial", 20), fg="white", bg="#6a0dad")
pregunta_label.pack(pady=10)

imagen_label = tk.Label(juego, bg="#6a0dad")
imagen_label.pack(pady=10)

opciones_frame = tk.Frame(juego, bg="#6a0dad")
opciones_frame.pack(pady=20)

juego.mainloop()
```
